# **Analysis of Udemy Course Dataset**

**Importing Libraries and Dataset**

```python
import pandas as pd
import numpy as np
```

```python
data=pd.read_csv('7. Udemy Courses.csv')
```

```python
data.head(3)
```

```python
data.size
```

**Data Cleaning**

```python
#Checking the amount of null values in the dataset
data.isnull().sum()
```

**Analysis of Data Set**

```python
data.columns
```

**What are the subjects on which Udemy offers courses?**

```python
data['subject'].unique()
```

**Which subject has the maximum number of courses?**

```python
data['subject'].value_counts()
```

**Show all the expert level courses**

```python
data[data.level=='Expert Level']
```

**Show the course name, price, num_reviews of course which is the most reviewed**

```python
maxi=data.num_reviews.max()
maxi
df=data[data.num_reviews == maxi]
cols=['course_title', 'num_reviews', 'price']
df=df[['course_title', 'num_reviews', 'price']]
df
```

**How many courses are Free?**

```python
data[data.price == 'Free'].count()
```

**Which are the top selling courses?**

```python
data.sort_values('num_subscribers', ascending=False)
```

**Show all Web Development courses with price equal to 100**

```python
data[(data.subject == 'Web Development') & (data.price == '100')]
```

**Shoq all courses related to python**

```python
data[data.course_title.str.contains('Web')]
```

**What are the courses published in 2014**

```python
data['published_timestamp']=pd.to_datetime(data.published_timestamp)
data['Year']=data['published_timestamp'].dt.year
data[data.Year == 2014]
```

**What are the maximum number of subscribers for each level of course**

```python
data.groupby('level')['num_subscribers'].max()
```
